```python
from selenium import webdriver
import time
import os
from selenium.webdriver.common.by import By
import pandas as pd
import threading
import requests
import datetime
```

```python
def scrapetext(prx, link,flag):
    
    arr = link.split("/")
    dir_name, f_name = '-'.join(arr[-4:-1]), arr[-1]+'.html'
#     print(dir_name)
#     print(f_name)
    if not os.path.exists(dir_name):
        os.mkdir(dir_name)

    try:
        html = requests.get(link,proxies=prx,timeout=2.5).text
        if len(html.split("\n")) == 1:
            print('Error')
            return link
        
        with open(dir_name+"/"+f_name,'w',encoding="utf-8") as f:
            f.write(html)
            
        return None
#             

    except requests.exceptions.ConnectTimeout:
        flag = True
        print(prx)
        with open(f'Error_{link.replace("/","_")}.txt','w+') as f:
            f.write(link)
            
        return link
            
```

```python
%%time
datetime.datetime.now()

list_of_Proxies = pd.read_csv("proxy_final.csv")
links = pd.read_csv("links2000s.csv")
initial_time = time.time()
downloaded_list = []
list_threads = []
failed_links = set()
global flag
flag = False
t = 0
global_counter = 0
while True:
    print(datetime.datetime.now())
    for prx in list_of_Proxies.proxy:
        prx = {"http":"http://"+prx}
        c = 0
        for link in links['Unnamed: 0'][t:]:
            print(prx, c)
            link = "https://api.parliament.uk"+link
            if link.endswith(".1"):
                link = link[:-2]
                
            
            t += 1
#             scrapetext(prx,link,flag)
#             scrapetext(prx,link,flag)
            list_threads.append(threading.Thread(target=scrapetext,args=(prx,link,flag)))
            list_threads[-1].start()
            
            
            previous_len_faild_links = len(failed_links)
            if c>3:
                print(datetime.datetime.now())
                for i in range(3):
                    failed_links.add(list_threads[i].join())                    
                
                if len(failed_links) != previous_len_faild_links:
                    with open('failed_links.txt','w') as f:
                        f.write(str(failed_links))
                        
                    previous_len_faild_links = len(failed_links)
                list_threads = []
                c = 0
                print(datetime.datetime.now())
                time.sleep(1)
                break
                
            c += 1
```
